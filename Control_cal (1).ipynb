```
import pandas as pd
from pyspc import * # importa todo de el módulo para gráficos de control de calidad
```

```
cerveza = pd.read_csv("https://raw.githubusercontent.com/rolex524/Datospilsenearlage/master/pilsenerlage.csv")
cerveza.set_index('Id', inplace = True) # establece el identificador de cada fila en Id
```

```
print(spc(cerveza)+xbar_sbar()+sbar()) # realiza los gráficos de control de la media y de la desviación estándar
# el proceso se halla bajo control
```
